```python
# Kth Largest Element

X = [10,5,2,6,3,7,12]

def kth_largest_elem(X,kth_larg):
    '''This function divides and eliminates elements to find the kth largest element in an array. It is efficient given it does not use sorting.
       Inputs: Array X where we seek the value, the value of k
       Output: the kth largest element that we require.'''
    k = len(X)-kth_larg
    # First I wrote the code for kth element (k) from index 0. 
    # This way, when we are asked for the kth largest element, I can take the kth element from the end.
    while (len(X)>0):  
        p = X[0]
        L = []
        R = []
        # Here we take out the first element and make it the the pivot element. 
        # We check the values greater and lesser than the pivot element and put them to the left and right of the pivot value
        for i in range(1,len(X)):
            if X[i] > p:
                R.append(X[i])
            elif X[i] < p:
                L.append(X[i])
        print(L,p,R)
        
        # We keep looping until we find our kth element, when we break out
        if len(L) == k:
            kth = p
            break
        # Here we check if the kth element is in the left list or the right list. 
        # If the element is in the right list, we remove the length of the left list and the pivot element to get the kth element
        elif len(L) > k:
            X = L
            k = k  
        elif len(L) < k:
            X = R
            k = k - len(L) - 1
    return kth    

print("Kth largest Element:", kth_largest_elem(X,1),"\n")
print("Kth largest Element:", kth_largest_elem(X,2),"\n")
print("Kth largest Element:", kth_largest_elem(X,3),"\n")
print("Kth largest Element:", kth_largest_elem(X,4),"\n")
print("Kth largest Element:", kth_largest_elem(X,5),"\n")
print("Kth largest Element:", kth_largest_elem(X,6),"\n")
print("Kth largest Element:", kth_largest_elem(X,7),"\n")
```
