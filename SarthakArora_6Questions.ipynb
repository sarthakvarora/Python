```python
def basket_balls(balls = 50, baskets = 50, simuls = 100):
    counters = []
    for j in range(simuls):
        basket = []
        for x in range(baskets):
            basket.append([])
        for i in range(balls):
            basket[random.randint(0,49)].append(1)
        counter = 0
        for y in basket:
            if len(y) == 0:
                counter += 1
        counters.append(counter)
    print(counters)    
    return baskets/(sum(counters)/len(counters))

print(basket_balls(50,50,100))        
```
