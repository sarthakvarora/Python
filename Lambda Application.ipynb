```python
def sum_squares(n,s=1):
    '''This function takes the values 'number' and 'step' as input,
       It squares all the numbers up till number (exclusive), with a default step value of 1, and gives out the sum of squares.
       e.g. Input: 10, 3
       Output: 1**2+4**2+7**2
    '''
    sums = 0
    for i in range(1,n,s):
        sums = sums + i*i
    return sums    
```

```python
sum_squares(100,2)
```

```python
sum_squares(100)
```

```python
# This function does the sum of squares up till a value, with a step function s which has a default value of 1

sum_sq = lambda x, s=1 : sum([x*x for x in range(1,x,s)])
```

```python
sum_sq(100,2)
```

```python
sum_sq(100)
```
