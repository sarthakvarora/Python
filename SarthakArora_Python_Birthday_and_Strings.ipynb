```python
def palindrome(text):
    '''This function will first remove the spaces from the text, and flip the string provided.
     Then it checks if it is the same as the initial string provided, to check if it is a palindrome
     Input: Text
     Output: Boolean - True if Palindrome, else False
     '''
    reverse = ""
    text2 = ""

    for alphabet in text:
        if alphabet.isalnum():
            text2 = text2 + alphabet.lower()

    for i in range(len(text2)):
        reverse = reverse + text2[-i-1]
    if reverse == text2:
        boolz = True
    else:
        boolz = False
    return boolz
```

```python
print(palindrome("Mr. Owl ate my metal worm!"))
```

```python
def substring_function(text,sub):
    '''This function will first remove the spaces from the text.
     Then it checks if the text contains the substring we are searching for
     Input: Text, substring to find
     Output: Boolean - True if Palindrome, else False
     '''

    text2 = ""
    sub = sub.lower()
    lensub = len(sub)

    for alphabet in text:
        if alphabet.isalnum():
            text2 = text2 + alphabet.lower()

    for i in range(len(text2)):
        subs = text2[i:i+lensub]
        if subs == sub:
            boolz = True
            break
        else:
            boolz = False
    return boolz        
```

```python
# Bday Paradox 0
import random

def bubbleSort(L):
    for a in range(0,len(L)):
        for b in range(0, len(L)-1-a):
            if L[b+1]<L[b]:
                L[b+1] , L[b] = L[b] , L[b+1]
    return L

def bday_paradox0(n):
    '''This function checks in n elements if we got a birthday twin and returns boolean value'''
    L = []
    L.append(random.randint(1,365))
    flag = False
    for i in range(n):
        L.append(random.randint(1,365))
        L = bubbleSort(L)
        print(L)
        for j in range(len(L)):
            if L[j] == L[j-1]:
                flag = True
                break     
    return flag

bday_paradox0(20)   
```

```python
# Bday Paradox 1

def bday_paradox1(n, sims):
    '''This will give for n users the probability that two people will have the same birthday. 
    I use multiple simulations of experiment to get probability'''
    boolz = []
    for runs in range(sims):
        L = []
        L.append(random.randint(1,365))
        flag = 0
        for i in range(n):
            L.append(random.randint(1,365))
            L = bubbleSort(L)
            for j in range(len(L)):
                if L[j] == L[j-1]:
                    flag = 1
                    break     
        boolz.append(flag)
    prob = sum(boolz)/len(boolz)
    return prob

print(bday_paradox1(23,100))
```

```python
# Bday Paradox 2

def bday_paradox2(probz):
    '''This will give out the number of users required to get two same birthdays with given probability'''
    L = []
    flag = 0
    prob = 1
    for i in range(1,100):
        L.append(random.randint(1,365))
        L = bubbleSort(L)
        # probability as each element is appended to list
        prob = prob*(1-i/365)
        proby = 1-prob
        if proby>probz:
            break
    return i           
    
print(bday_paradox2(0.8))    
```

The problem with just a single run is that it gives a test run, and hence a theoretical probability. With just one run, we might have a case where we might get a twin with ten users, or with fifty users. And that won't be the accurate representation of the number of runs it takes on average.

Hence we run multiple simulations and take their average to get a better idea of the actual number of users required.

```python
# Bday Paradox 3

def bday_paradox3(prob, sims):
    ''''''
    Main = []
    probu = []
    for i in range(1,101):
        boolz = []
        for runs in range(sims):
            L = []
            L.append(random.randint(1,365))
            flag = 0
            for k in range(i):
                L.append(random.randint(1,365))
                L = bubbleSort(L)
                for j in range(len(L)):
                    if L[j] == L[j-1]:
                        flag = 1
                        break     
            boolz.append(flag)
        Main.append(boolz)
    for k in range(len(Main)):
        probu.append(sum(Main[k])/len(Main[k]))
    for l in range(len(probu)):
        if probu[l] > prob:
            indx = l+1
            break
    return indx
    
print(bday_paradox3(0.8,50))
```
